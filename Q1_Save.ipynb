```python
import pandas as pd
```

```python
df=pd.read_excel('Housing (1).csv.xlsx')
```

```python
df
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
for i in ['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']:
    df[i]=label_encoder.fit_transform(df[i])
 
```

```python
df
```

```python
target=df['price']
```

```python
df.drop('furnishingstatus',1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df,target,test_size=0.2,random_state=109)
```

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train,y_train)
```

```python
y_pred=regr.predict(X_test)
print(y_pred)
```

```python
print(y_test)
```

```python
regr.score(X_test,y_test)
```

```python
print('theta 0',regr.intercept_)
print('theta 1',regr.coef_)
```

```python
# Save Model Using pickle
import pickle
filename = 'finalized_model.sav'
pickle.dump(regr, open(filename, 'wb'))
```
